```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### 1-VAZIFA. Avvalgi [amaliyotda](https://colab.research.google.com/drive/1jzhJ1gN6msILYOHaG2fhBBvUGaa4tcLf?usp=sharing) saqlangan faylni yangi `df`ga o'qing

```python
#JAVOBINGIZNI SHU YERGA YOZING
!wget  https://en.tutiempo.net/climate/ws-384570.html
```

```python
df=pd.read_html("/content/ws-384570.html",index_col=0)
df=df[3]
```

```python
df=df[df.index>1990]
```

```python
df=df.replace("-",np.nan)
```

```python
df=df.fillna(method='ffill')
```

```python
df=df.astype(dtype=float)
```

```python
df.info()
```

```python
#@title Misol:
df.describe()
del df["PP"]
```

### 2-VAZIFA. `df` barcha ustunlarini chiziqli grafik ko'rinishida chiqaring

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(16,8))
sns.lineplot(data=df,alpha=1,linewidth=2)
plt.title("Toshkentdagi yillik ob-havo")
plt.xlabel("yillar")
plt.ylabel("harorat")
plt.legend(loc=1)
plt.show()
```

```python
#@title Javob
plt.figure(figsize=(16,6))
sns.lineplot(data=df)
plt.title("Toshkent shahri yillik ob-havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.show()
```

### 5-VAZIFA. Yillik o'rtacha shamol tezligini ustunli grafik ko'rinishida chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(16,8))
sns.barplot(x=df.index,y=df['V'])
plt.xticks(rotation=90)
plt.ylabel("O'rtacha shamol tezligi")
plt.xlabel("Yillar")
plt.show
```

```python
df2=df.[['RA', 'SN', 'TS', 'FG', 'TN']].T
```

### 6-VAZIFA. Yomg'rili, qorli, bo'ronli, tumanli va do'l yoqqan kunlar sonini heatmap ko'rinishida chiqaring.

```python
#JAVOBINGIZNI SHU YERGA YOZING
plt.figure(figsize=(30,3))
sns.heatmap(data=df2.rename(index={"RA":"Yomg'irli", "SN":"Qorli", "TS":"bo'ronli","FG":"tumanli","TN":"do'l"}),square=True,cmap="YlGnBu",linewidth=1,alpha=1)
plt.yticks(rotation=0)
plt.xticks(rotation=90)
plt.show()
```

```python
#@title Kutilgan natija:
plt.figure(figsize=(30,5))
sns.heatmap(data=df[['RA','SN','TS','FG','GR']].T, cmap="YlGnBu", linewidths=1, square=True)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
yticks_labels = ["Yomg'ir","Qor","Bo'ron","Tuman","Do'l"]
plt.yticks(np.arange(5)+0.5, labels=yticks_labels)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
