```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
```

```python
df.head()
```

```python
df=df[df["year"]>2000]
plt.figure(figsize=(16,6))
sns.scatterplot(data=df,x=df["year"],y=df['fuelType'])
plt.show()
```

```python
df=df[df["year"]>2000]
plt.figure(figsize=(16,6))
sns.regplot(data=df,x=df["year"],y=df['mpg'])
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(18,9))
sns.scatterplot(data=df,x=df["mileage"],y=df['price'],hue='transmission',style='fuelType',s=100)

plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.swarmplot(data=df,x=df["transmission"],y=df['price'])
```
